```python
import zipfile
import pandas as pd

zip_path = 'dataset/Alzheimer_s_Disease_and_Healthy_Aging_Data_20241126.zip'

with zipfile.ZipFile(zip_path, 'r') as z:
    print("Files in zip:", z.namelist())
    csv_file = z.namelist()[0]
    with z.open(csv_file) as f:
        df = pd.read_csv(f)
df.head()
```
